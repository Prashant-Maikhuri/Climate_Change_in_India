```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = pd.read_csv("temp_change.csv")
```

```python
data1
```

```python
x=data1["Months"]
y1=data1["Y2015"]
y2=data1["Y2016"]
y3=data1["Y2017"]
y4=data1["Y2018"]
y5=data1["Y2019"]
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
plt.plot(x,y1,label="2015")
plt.plot(x,y2,label="2016")
plt.plot(x,y3,label="2017")
plt.plot(x,y4,label="2018")
plt.plot(x,y5,label = "2019")
plt.legend()
plt.xlabel("Months")
plt.title("Climate Change")
```
